```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dados = {
    "Modelo": ["XGBoost","XGBoost","XGBoost",
               "RF","RF","RF",
               "SVM","SVM","SVM"],
    "MOTIF": ["P_LOOP","GLPL","MHD"]*3,
    "arabidopsis":[20,14,15,20,13,14,18,8,11],
    "tuberosum":[5,4,1,6,5,1,7,6,2],
    "physalis":[12,9,3,20,16,3,11,10,4]
}

df = pd.DataFrame(dados)

for especie in ["arabidopsis","tuberosum","physalis"]:

    tabela = df.pivot(
        index="Modelo",
        columns="MOTIF",
        values=especie
    )

    plt.figure(figsize=(6,4))

    sns.heatmap(
        tabela,
        annot=True,
        cmap="viridis",
        fmt="d"
    )

    #plt.title(f"Detecção de Motifs - {especie}")
    plt.tight_layout()
    plt.savefig(
            f"heatmap_{especie}.png",
            bbox_inches="tight"
        )
    plt.show()
```
